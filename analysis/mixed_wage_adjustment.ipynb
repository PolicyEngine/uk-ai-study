## tl;dr

At a fixed gross earnings loss, shifting adjustment from wage cuts to displacement lowers the fiscal cost but raises poverty and inequality. Adverse survivor wages are a separate, large fiscal sensitivity.

## Context & Methods

This companion reads the committed outputs of `sensitivity_mixed_wage_adjustment.py`. The source is the FRS 2024-25 PolicyEngine dataset, simulated for 2026 at seed 0.

### Key Assumptions

For seed 0, the mixed grid holds gross pre-uplift earnings loss fixed at the central 7% employee-displacement draw's realised loss. Lambda scales the displacement rate and C-AIOE-graded cuts among survivors fill the gap. The survivor-wage grid fixes the central displacement draw and varies only the complementarity-graded wage parameter.

```python
from pathlib import Path
import pandas as pd

root = Path.cwd().parent if Path.cwd().name == 'analysis' else Path.cwd()
results = root / 'results' / 'robustness'
mixed = pd.read_csv(results / 'mixed_adjustment.csv')
wages = pd.read_csv(results / 'survivor_wage_grid.csv')
```

## Data

The two grids and the paper-facing metrics are shown below.

```python
mixed[['lambda_displacement_share', 'gross_earnings_loss_pct', 'displaced_weighted_m', 'exchequer_cost_bn', 'poverty_change_bhc_pp', 'gini_change_pp']].round(2)
```

```python
wages[['survivor_wage_effect_pct', 'exchequer_cost_bn', 'poverty_change_bhc_pp', 'gini_change_pp']].round(2)
```

## Results

Invariant and endpoint checks verify the common gross loss and reproduce the central displacement scenario.

```python
assert mixed.loc[0, 'displaced_weighted_m'] == 0
assert mixed['gross_earnings_loss_pct'].max() - mixed['gross_earnings_loss_pct'].min() < 1e-10
assert abs(mixed.loc[0, 'exchequer_cost_bn'] - 26.566918144) < 1e-8
assert abs(mixed.loc[4, 'exchequer_cost_bn'] - 18.422136832) < 1e-8
central = wages.loc[wages['survivor_wage_effect_pct'].sub(2.6).abs().idxmin()]
assert abs(central['exchequer_cost_bn'] - mixed.loc[4, 'exchequer_cost_bn']) < 1e-8
'Common-loss invariant and central-scenario reconciliation passed'
```

## Takeaways

With gross pre-uplift earnings loss fixed at 7.765% of the baseline wage bill, the fiscal cost falls from £26.6bn to £18.4bn as adjustment shifts from wage cuts to displacement. Poverty rises from 0.10 to 1.87 percentage points and the Gini effect changes from -0.44 to +1.10 points. At fixed central displacement, replacing the +2.6% survivor uplift with a -2.6% wage effect raises the fiscal cost from £18.4bn to £45.3bn. These are author-designed reduced-form sensitivities, not estimated equilibrium responses.
